# 01 - Entendimento dos dados

Objetivo: entender a base de churn, o alvo do problema e os principais sinais antes da modelagem.


```python
# Load libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

DATA_PATH = '../data/raw/telco_churn_raw.csv'
df = pd.read_csv(DATA_PATH)
df.shape
```

```python
# Inspect first records
df.head()
```

```python
# Create metadata table
metadata = pd.DataFrame({
    'variavel': df.columns,
    'tipo': df.dtypes.astype(str),
    'n_nulos': df.isna().sum().values,
    'pct_nulos': (df.isna().mean().values * 100).round(2),
    'cardinalidade': [df[c].nunique() for c in df.columns]
})
metadata
```

```python
# Check target balance
churn_dist = df['Churn'].value_counts(normalize=True).rename('proporcao').reset_index()
churn_dist
```

```python
# Plot target distribution
ax = df['Churn'].value_counts().plot(kind='bar', color=['#2f6f9f', '#cc4125'], figsize=(6, 4))
ax.set_title('Distribuicao do churn')
ax.set_xlabel('Churn')
ax.set_ylabel('Clientes')
plt.tight_layout()
plt.show()
```

```python
# Analyze churn by contract type
contract_churn = df.groupby('Contract')['Churn'].apply(lambda s: (s == 'Yes').mean()).sort_values(ascending=False)
contract_churn
```
